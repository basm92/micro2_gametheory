```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp

import matplotlib
```

```python
# Parameters
b = 1
a = 10
c = 1

root = 0
```

```python
def plot_pi_1(q2, q3):
    root = (a - b*q2 - b*q3 - c)/b
    q1 = np.array(np.linspace(0.0, root, num=100))
    pi = -b * q1**2 + (a-c-b*(q2+q3))*q1
    return plt.plot(q1, pi)
```

```python
plot_pi_1(2,2)
plot_pi_1(3,3)
plot_pi_1(4,4)
```

```python
# player 1 plays maximum (a-c)/2b
# What does player 2 then do?
# q1 = (a-c)/2*b

def plot_pi_2(q1, q3):
    root = (a - b*q1 - b*q3 - c)/b
    q2 = np.array(np.linspace(0.0, root, num=100))
    pi = -b * q2**2 + (a-c-b*(q1+q3))*q2
    return plt.plot(q2, pi)
```

```python
plot_pi_2((a-c)/(2*b), 0)
plot_pi_2((a-c)/(2*b), 2)
```
